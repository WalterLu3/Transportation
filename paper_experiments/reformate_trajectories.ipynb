```python
# change file
import csv
import numpy as np
import pickle
```

```python
with open("trajectories_500_large_randomness.csv",'r') as f:
    read = csv.reader(f)
    read = list(read)
read = read[1:]
```

```python
result = [['road','loopNum']]

for e in read:
    #print(e[1][4:])
    result.append([e[0],'l'+str((int(e[1][4:])-1)*100 + int(e[2][1:]))])
```

```python
with open("trajectories_500_L.csv",'w') as f:
    writer = csv.writer(f)
    writer.writerows(result)
```
